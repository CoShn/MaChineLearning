```python
import pandas as pd
import matplotlib.pyplot as plt

fakultas = ["Bisnis", "D3 Perhotelan", "ICT", "Ilmu Komunikasi", "Seni dan Desain"]
jumlah_mahasiswa = [260, 28, 284, 465, 735]
akreditasi = ["A", "A", "B", "A", "A"]
Colour = ["RED", "BROWN", "GREEN", "BLUE", "PURPLE"]
info_mahasiswa = pd.DataFrame({
    "fakultas": fakultas,
    "jumlah_mahasiswa": jumlah_mahasiswa,
    "akreditasi": akreditasi,
})

fig, ax = plt.subplots()
ax.bar(info_mahasiswa["fakultas"], info_mahasiswa["jumlah_mahasiswa"],
       width=0.8,color=Colour)
ax.set_ylabel("Jumlah Mahasiswa", fontdict={'fontsize': 12, 'fontweight': 'bold', 'color': 'black'})
ax.set_xlabel("Fakultas", fontdict={'fontsize': 12, 'fontweight': 'bold', 'color': 'black'})
ax.set_title("Jumlah Mahasiswa per Fakultas", fontdict={'fontsize': 12, 'fontweight': 'bold', 'color': 'black'})
ax.tick_params(axis='x', labelrotation=0, labelsize=7.5)
```
